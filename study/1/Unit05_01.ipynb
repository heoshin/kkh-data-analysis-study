```python
import csv
import matplotlib.pyplot as plt
# 2003.01.01~2022.05.31 데이터 사용
f = open('My_birthday.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[4] != '' :
        result.append(float(row[4]))
plt.plot(result, 'r')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
# 2003.01.01~2022.05.31 데이터 사용
f = open('My_birthday.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[4] != '' :
        if row[2].split('-')[1] == '05' and row[2].split('-')[2] == '07':
            result.append(float(row[4]))
plt.plot(result, 'hotpink')
plt.show()
```
