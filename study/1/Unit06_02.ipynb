```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[4] != '' :
        result.append(float(row[4]))
        
plt.hist(result, bins=100, color='r')
plt.show()
```

```python
#5월 데이터 뽑기
import csv
import matplotlib.pyplot as plt

f = open('My_birthday.csv')
data = csv.reader(f)
next(data)
may = []

for row in data :
    month = row[2].split('-')[1]
    if row[4] != '' :
        if month == '05' :
            print(row)
            may.append(float(row[4]))
        
plt.hist(may, bins=100, color='r')
plt.show()
```
